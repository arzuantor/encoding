```python
ds=df[["Segment","City","Sub-Category","Ship Mode","Region","Category"]]
```

```python

df1 = ds.copy()
df2 = ds.copy()
df3 = ds.copy()
df4 = ds.copy()
df5 = ds.copy()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

#LableEncoding

```python
le=LabelEncoder()
df1['Category']=le.fit_transform(df1['Category'])
df1
```

```python
df1['Category'].value_counts()
```

#ordinal encoder


```python
ord=OrdinalEncoder()
```

```python
df1["Ship Mode"]=ord.fit_transform(df1[["Ship Mode"]])
```

```python
df1.head(1)
```

#one hot encoding

```python
ohe=OneHotEncoder(sparse_output=False)
```

```python
df1.head(1)
```

```python
encode=ohe.fit_transform(df1[['Region']])
```

```python
encode
```

```python
df1_new=pd.DataFrame(encode,columns=ohe.get_feature_names_out(['Region']))
df1_new.head(1)
```

```python
df1=df1.drop(["Region"],axis=1)
```

```python
df1
```

```python
df1_update=pd.concat([df1,df1_new],axis=1)
```

```python
df1_update['Segment'].value_counts()
```

```python
df1_update
```

#create dummy variable

```python
dummy=pd.get_dummies(df1_update['Segment'],drop_first=True).astype("int64")
```

```python
dummy
```

```python
df1_update=df1_update.drop(["Segment"],axis=1)
```

```python
df1_update.head()
```

```python
df1_update=pd.concat([df1_update,dummy],axis=1)
```
