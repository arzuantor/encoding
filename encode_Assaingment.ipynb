#using loop for on hot encoding

```python
for col in ds.columns:
    if is_numeric_dtype(ds[col]):
        continue
    else:
        dummy = pd.get_dummies(ds[col], drop_first=True).astype("int64")
        ds = ds.drop([col], axis=1)
        ds = pd.concat([ds, dummy], axis=1)
```
